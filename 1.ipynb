```python
import pandas as pd

demographics = "C:/Beam/data_pack/demographics.csv"
cc = "C:/Beam/data_pack/cc.csv"
kPlus = "C:/Beam/data_pack/kplus.csv"

demographics = pd.read_csv(demographics)
cc = pd.read_csv(cc)
kPlus = pd.read_csv(kPlus)
```

```python
# Credit card
unique_name = cc["cc_no"].unique()    # unique name of credit_card number
cc_df = pd.DataFrame(cc["cc_no"])     # Create dataframe
each_card_spending = [None]*len(unique_name)

for card in unique_name:
    monthly_spending = [0,0,0,0,0,0]    # Montly spending of each card
    indices = cc_df[cc_df == card].stack().index.tolist()
    for i in indices:
        i = i[0]
        month = int(cc["pos_dt"][i][6])
        monthly_spending[month-1] += cc["cc_txn_amt"][i]
    print("card ", card, " spend ", monthly_spending)
    each_card_spending[unique_name.tolist().index(card)] = monthly_spending

each_card_spending
```

```python
# Debit card
unique_name = demographics["id"].unique()  
kPlus_df = pd.DataFrame(kPlus["id"])
each_id_spending = [None]*len(unique_name)

for name in unique_name:
    monthly_spending = [0,0,0,0,0,0]
    indicies = kPlus_df[kPlus==name].stack().index.tolist()
    for i in indicies:
        i = i[0]
        month = int(kPlus["sunday"][i][6])
        monthly_spending[month-1] += kPlus["kp_txn_amt"][i]
    print("name id ", name, " spends ", monthly_spending)
    each_id_spending[unique_name.tolist().index(name)] = monthly_spending

each_id_spending
```

```python
data = {
    "Credit": each_card_spending,
    "Debit": each_id_spending
}
df_data = pd.DataFrame.from_dict(data)

export_csv = df_data.to_csv (r'C:\Beam\data_pack\Credit-debit.csv', index = None, header=True)
```
